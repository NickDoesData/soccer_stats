# Passing Stats and Total Points Correlation Analysis

```python
import pandas as pd
```

Stats have been downloaded from whoscored.com

```python
df = pd.read_csv('passing_stats.csv')
```

```python
df['LB_pct'] = df.AccLB / (df.AccLB + df.InAccLB)
df['SP_pct'] = df.AccSP / (df.AccSP + df.InAccSP)
```

```python
df[['AccLB', 'LB_pct','AccSP','SP_pct']].corrwith(df['Pts'])
```

LB pct has a surprising level of correlation to points earned in the table. The amount of successful short passes has the highest correlation, which makes sense as these teams probably have the highest amount of possession.

```python
df[['LB_pct']].corrwith(df['SP_pct'])
```

Long ball accuracy is very strongly correlated with short passing accuracy. Those who are accurate passers are accurate regardless of distance.

```python
print 'Accurate LB Table'
df_LB = df[['Team','AccLB','LB_pct' ,'Pts']].sort_values('AccLB', ascending=False).reset_index(drop=True)
df_LB.index += 1
df_LB
```

```python
print 'LB % Table'
df_LB_pct = df[['Team','AccLB','LB_pct' ,'Pts']].sort_values('LB_pct', ascending=False).reset_index(drop=True)
df_LB_pct.index += 1
df_LB_pct
```

```python
print 'Short Pass Table'
df_SP = df[['Team','AccSP', 'SP_pct','Pts']].sort_values('AccSP', ascending = False).reset_index(drop=True)
df_SP.index += 1
df_SP
```

```python
print 'SP % Table'
df_SP_pct = df[['Team','AccSP','SP_pct' ,'Pts']].sort_values('SP_pct', ascending=False).reset_index(drop=True)
df_SP_pct.index += 1
df_SP_pct
```
